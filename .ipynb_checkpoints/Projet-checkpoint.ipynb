```python
# 1.1 Initialisation de l'envrionnement
import pandas as pan
import webbrowser as web
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1.2 - Variable de données
data_table = pan.read_csv("data.csv")
```

```python
# 2.1 - Création d'une page de vue html
html = data_table[0:100].to_html()
```

```python
# 2.2 - Ouverture dans un navigateur
with open("data.html", "w", encoding='utf-8') as f:
    f.write(html)
full_fillname = os.path.abspath("data.html")
web.open("file://{}".format(full_fillname))
```

```python
# 3 - Variables : types et attributs
print('Type :\n')
print(data_table.dtypes)
print('\nAttributs :\n')
print(data_table.columns)
```

```python
# 4 - Histogramme
sns.distplot(data_table['Overall'], bins=49)
plt.show()
```

```python
# 4.3 - Quelques valeures de la valeures globales
print("Moyenne :", data_table['Overall'].mean())
print("Médiane :", data_table['Overall'].median())
print("Écart-type :", data_table['Overall'].std())
print("Variance :", data_table['Overall'].var())
print("Minimum :", data_table['Overall'].min())
print("Maximum :", data_table['Overall'].max()) 
```

```python
# 5 - Description de Nationalité
print("\nNationality description :")
print(data_table['Nationality'].describe())
```

```python
# 6 - Countplots
nationality_data = data_table[['Nationality', 'Potential']] # Initialisation de nationality_data

nationality_data = nationality_data[(nationality_data['Nationality'] == 'France') |
                                  (nationality_data['Nationality'] == 'Italy') |
                                  (nationality_data['Nationality'] == 'Spain') |
                                  (nationality_data['Nationality'] == 'Brazil') |
                                  (nationality_data['Nationality'] == 'Argentina') |
                                  (nationality_data['Nationality'] == 'Germany') |
                                  (nationality_data['Nationality'] == 'England')]

sns.barplot(x='Nationality', y='Potential', capsize=0.5, ci='sd', data=nationality_data) # Paramètres d'affichage du countplot
plt.show()
```

```python
# 7 - Scatterplot
sns.scatterplot(x='Nationality', y='Potential', data=nationality_data)
plt.show()
```
